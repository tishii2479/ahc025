```python
import numpy as np
import matplotlib.pyplot as plt

# パラメータ設定
lambda_param = 1e-5

# 指数分布のランダムサンプルを生成
sample_size = 1000
exponential_data = np.random.exponential(scale=1/lambda_param, size=sample_size)

# 理論的な指数分布の確率密度関数を描画
x = np.linspace(0, max(exponential_data), 1000)
pdf = lambda_param * np.exp(-lambda_param * x)
plt.plot(x, pdf, 'r', label='Theoretical PDF')

plt.xlabel('value')
plt.ylabel('p')
plt.legend()

plt.title('exp distribution')
plt.show()
```
