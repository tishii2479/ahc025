```python
import matplotlib.pyplot as plt
```

```python
a = []
b = []

with open("a.txt", "r") as f:
    for line in f.readlines():
        v = line.split()
        if v[0] == "a:":
            a.append(int(v[1]))
        elif v[0] == "b:":
            b.append(int(v[1]))
```

```python
plt.hist(a)
```

```python
plt.hist(b)
```
